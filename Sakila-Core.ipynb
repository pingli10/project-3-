```python
from sqlalchemy import create_engine


```

```python
# Create connection string using credentials following this format
# connection = "dialect+driver://username:password@host:port/database"
connection = "mysql+pymysql://root:mypassword@localhost/sakila"

```

```python
engine = create_engine(connection)

```

```python
q = """SELECT customer.first_name,customer.last_name,customer.email,address.address,address.district
FROM customer
JOIN address ON customer.address_id = address.address_id
WHERE address.city_id = 312"""
pd.read_sql(q, engine)
```

```python
q = """SELECT film.title,film.description,film.release_year,film.rating,film.special_features
FROM film
JOIN category ON film.film_id = category.category_id
WHERE category.name = 'Comedy'"""
pd.read_sql(q, engine)
```

```python
q = """SELECT actor.last_name,film.title,film.release_year
FROM film_actor INNER JOIN actor
ON film_actor.actor_id = actor.actor_id
INNER JOIN film 
ON film_actor.film_id = film.film_id
WHERE actor.first_name = 'Johnny'
AND actor.last_name = 'Lollobrigida'"""
pd.read_sql(q, engine)
```

```python
q = """
SELECT actor.first_name,actor.last_name
FROM film_actor INNER JOIN actor
ON film_actor.actor_id = actor.actor_id
INNER JOIN film 
ON film_actor.film_id = film.film_id
WHERE film.title = 'Bingo Talented'
"""
pd.read_sql(q, engine)
```

```python
q = """
SELECT customer_id
FROM payment
WHERE amount > 4.20 * 2"""
pd.read_sql(q, engine)
```

```python
q = """SELECT customer.first_name,customer.last_name,count(amount) as num_payments
FROM payment
JOIN customer ON payment.payment_id = customer.customer_id
GROUP BY customer.customer_id
ORDER BY num_payments DESC
LIMIT 5
"""
pd.read_sql(q, engine)
```
